```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
df = pd.read_csv(r'Shakespeare_data.csv')
df
```

```python
#The next thing to do is to start cleaning and interpreting the data
#First, the stage direction lines will be stricken from the dataset to clean the data
#Next, the dataline column will be stricken from the dataset to keep the model from getting confused from pointless information
#Then, the ActSceneLine column will be split with an ActScene column and a Line column. The line column will be deleted as well, as the PlayerLineNumber column is a better descriptor
#Finally, the Player and Play columns will be encoded to ensure easier reading from the program
```

```python
Clean_data = df[pd.notna(df['Player'])]
Clean_data = df[pd.notna(df['ActSceneLine'])]
Clean_data = Clean_data.drop(['Dataline'], axis = 1)
all_plays = df['Play'].unique()
play_list = all_plays.tolist()
#print(play_list) #obtains play titles

Clean_data[['ActScene','Line']] = Clean_data['ActSceneLine'].str.rsplit('.', 1, expand=True).astype('float') #get two different columns with the Act + Scene and the Lines
Clean_data = Clean_data.drop(['Line'], axis = 1)
Clean_data = Clean_data.drop(['ActSceneLine'], axis =1)
Clean_data = Clean_data.drop(['PlayerLine'], axis =1)


le = LabelEncoder()
le.fit(Clean_data['Play'])
Clean_data['Play'] = le.transform(Clean_data['Play'])

# le.fit(Clean_data['Player']) #This was the final part of the code that would not fix itself. There was an unknown error that did not allow me to proceed
# Clean_data['Player'] = le.transform(Clean_data['Player'])
Clean_data
```

```python
#There were initially plans to allow for another feature that noted which characters would show up in each scenes, but this was cut due to time constraints
```

```python
#Now, it is time to use Machine learning techniques to analyze the altered data
#The first method that I tried was the decision tree model
```

```python
x = Clean_data.drop('Player', axis = 1)
y = Clean_data.Player

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25)

training_model = DecisionTreeClassifier()
training_model.fit(x_train, y_train)

```

```python
y_pred = training_model.predict(x_test)

test_score = metrics.accuracy_score(y_test, y_pred)
```
